```python
import numpy as np
arr=([1,2,3,6,4,5])
print("Reverse array:", arr[::-1])
```

```python
arr1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])
print("Flatten:", arr1.flatten())
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])
print("comparing: ", np.array_equal(arr1, arr2))
```

```python
x = np.array([1, 2, 3, 4, 5, 2, 1, 2, 1, 1])
y = np.array([1, 1, 1, 2, 3, 2, 4, 4, 3, 3, 1])
print("Most frequent in x:", np.bincount(x).argmax())
print( np.where(x == np.bincount(x).argmax()))
print("Most frequent in y:", np.bincount(y).argmax())
print( np.where(y == np.bincount(y).argmax()))
```

```python
gfg = np.matrix([[4, 1,9], [12, 3, 1], [4, 5, 6]])
print("Sum of all elements:", gfg.sum())
print("Sum rowwise:", gfg.sum(axis=1))
print("Sum column-wise:", gfg.sum(axis=0))
```

```python
n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])
print("Sum of diagonal:", np.trace(n_array))
print("Eigenvalues:", np.linalg.eig(n_array))
print("Inverse:", np.linalg.inv(n_array))
print("Determinant:", np.linalg.det(n_array))


```

```python
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])
print("Matrix product: ", np.dot(p, q))
print("Covariance: ", np.cov(p, q))
```

```python
x = np.array([[2,3,4],[3,2,9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
print("Inner product:", np.inner(x, y))
print("Outer product:", np.outer(x, y))
print("Cartesian product: ", np.array(np.meshgrid(x, y)).T.reshape(-1, 2))
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])
print("Absolute values:\n", np.abs(array))
print("ROWS:" )
print("25th percentile:", np.percentile(array, 25, axis=1))
print("50th percentile:", np.percentile(array, 50, axis=1))
print("75th percentile:", np.percentile(array, 75, axis=1))
print("Mean :", np.mean(array, axis=1))
print("Median:", np.median(array, axis=1))
print("Std deviation :", np.std(array, axis=1))
print("Across Coloumns: ")
print("25th percentile:", np.percentile(array, 25, axis=0))
print("50th percentile:", np.percentile(array, 50, axis=0))
print("75th percentile:", np.percentile(array, 75, axis=0))
print("Mean (columns):", np.mean(array, axis=0))
print("Median (columns):", np.median(array, axis=0))
print("Std deviation (columns):", np.std(array, axis=0))
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
print("Floor:", np.floor(a))
print("Ceiling:", np.ceil(a))
print("Truncate:", np.trunc(a))
print("Round: ",np.round(a))
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("Sorted array:", np.sort(array))
print("Indices sorted:", np.argsort(array))
print("4 smallest:", np.sort(array)[:4])
print("5 largest:", np.sort(array)[-5:])
```

```python
array2 = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
integer = array2[np.floor(array2) == array2]
floats = array2[np.floor(array2) != array2]
print("Integers only:", integer)
print("Floats only:", floats)
```
