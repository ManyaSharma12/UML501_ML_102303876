```python
from bs4 import BeautifulSoup
import requests
url="https://books.toscrape.com/"
page=requests.get(url)
soup=BeautifulSoup(page.content,"html.parser")
```

```python
	
books=soup.find_all("article",class_="product_pod")
for book in books:
    title=book.h3.a["title"]
    print(title)
```

```python
for book in books:
    title=book.h3.a["title"]
    price=book.find("p",class_="price_color").text
    print(title," -> ",price)
```

```python
for book in books:
    title=book.h3.a["title"]
    stock=book.find("p",class_="instock availability").text.strip()
    print(title,"->",stock)
```

```python
i=1
data=[]
for book in books:
    title=book.h3.a["title"]
    price=book.find("p",class_="price_color").text
    stock=book.find("p",class_="instock availability").get_text(strip=True)
    rating_tag=book.find("p",class_="star-rating")
    ratings=rating_tag["class"][1]
    print(f" {i}) {title} → {price} → {stock} → {ratings} Stars")
    i=i+1
    data.append([title,price,stock,ratings])
```

```python
import pandas as pd
df=pd.DataFrame(data,columns=["Title","Price","Stock","Ratings"])
df.to_csv("books.csv",index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/chart/top/")
```

```python
movies=driver.find_elements(By.CSS_SELECTOR,"ul.ipc-metadata-list ipc-metadata-list--dividers-between sc-2b8fdbce-0 emPbxy compact-list-view ipc-metadata-list--base")
```

```python
data=[]
for movie in movies[:10]:
    title_block = movie.find_element(By.CSS_SELECTOR, "h3.ipc-title__text").text
    rank, title = title_block.split(". ", 1)
    year = movie.find_element(By.CSS_SELECTOR, "span.sc-15ac7568-7.cCsint.cli-title-metadata-item").text
    year = year.strip("()")
    rating = movie.find_element(By.CSS_SELECTOR, "span.ipc-rating-star--rating").text
    print(rank,title,"->",year,"->",rating)
    data.append([int(rank), title, int(year), float(rating)])
```
